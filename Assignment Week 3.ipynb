#### Part III: Cluster the neighborhood Toronto

We need to first create a dataframe that contains Toronto in Borough name

```python
df3=df2[df2['Borough'].str.contains('Toronto')] 
```

```python
df3.head()
```

```python
df3.reset_index(drop=True) #redo the index
```

Time to import all the necessary packages

```python
from sklearn.cluster import KMeans
import folium
```

We want to generate dots based on latitude and longitude

```python
df4=df3[["Latitude","Longitude"]].reset_index(drop=True)
df4
```

```python
# set number of clusters
kclusters = 5


# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0,n_init = 12).fit(df4) #repeat 12 times to get a better idea

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
labels=kmeans.labels_ #give the labels a name
```

```python
predict=kmeans.predict(df4)
df3['labels'] = pd.Series(predict, index=df3.index)
df3.head()
```

Now it is time to generate a map

```python
latitude=43.67
longitude=-79.39 #pin point to somewhere near Toronto
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df3['Latitude'], df3['Longitude'], df3['Neighborhood'], df3['labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
